## 2. Data Cleaning

### 2.1. Data types

#### 1. Do the height and weight columns have the right data types?

```python
height_dt = df['Height'].dtypes
weight_dt = df['Weight'].dtypes
print(f"Weight column is {weight_dt} type and height column is {height_dt} type.")
```

### 2.2 Changing data types

```python
# dtype is an 'Object', these columns should be converted to integers

def convert_to_int(x):
    try:
        if x.endswith("kg"):
            #removing the unit 'kg'
            x = x.replace("kg", "")
        elif x.endswith("cm"):
            #removing the unit 'cm'
            x = x.replace("cm", "")
        #converting x to an integer
        x = int(x)
    except ValueError:
        #if there are any other unit or misspeling in collected data, it will replace it with average height/weight
        if x.endswith("cm"):
            return 175
        elif x.endswith("kg"):
            return 70
    return x



#applying the function
df['Height'] = df['Height'].apply(lambda x: convert_to_int(x))
df['Weight'] = df['Weight'].apply(lambda x: convert_to_int(x))

df
```

### 2.3. Renaming columns

```python
df.rename(columns={'Height': 'Height (cm)'}, inplace=True)
df.rename(columns={'Weight': 'Weight (kg)'}, inplace=True)
df
```

### 2.4 Separating a column

#### 2. Can the "joined" column be separated into year, month and day columns?

```python
#getting a year column
def get_year(year):
    return year.split(',')[1]

#getting a month with a day column
def get_month(month):
    return month.split(',')[0]


df['Year'] = df['Joined'].apply(lambda x: get_year(x))
df['Month'] = df['Joined'].apply(lambda x: get_month(x))
df
```

```python
#getting a day in a month
df[['Month','Day']] = df['Month'].str.split(' ', expand=True)
df
```

### 2.5. Transforming a column

#### 3. Is it possible to clear and convert the "Value", "Salary" and "Release Clause" columns into integer columns?

```python
#Value,Wage,Release Clause ---> Int
def convert_to_integer(x):
    #removeing the euro sign
    x = x.replace("€", "")
    #checking if the string ends with M or K
    if x.endswith("M"):
        #removing the letter M
        x = x.replace("M", "")
        #converting the string to a float and multiply by 1000000
        x = float(x) * 1000000
    elif x.endswith("K"):
        #removing the letter K
        x = x.replace("K", "")
        #converting the string to a float and multiply by 1000
        x = float(x) * 1000
    else:
        #converting the string to a float
        x = float(x)
    #converting the float to an integer and return it
    x = int(x)
    return x
#applying the function
df['Value'] = df['Value'].apply(lambda x: convert_to_integer(x))
df['Wage'] = df['Wage'].apply(lambda x: convert_to_integer(x))
df['Release Clause'] = df['Release Clause'].apply(lambda x: convert_to_integer(x))

#renaming the columns so that we know what currency it is
df.rename(columns={'Value': 'Value (€)'}, inplace=True)
df.rename(columns={'Wage': 'Wage (€)'}, inplace=True)
df.rename(columns={'Release Clause': 'Release Clause (€)'}, inplace=True)
df
```

### 2.6. Removing from a column

#### 4. How can you remove the newline characters from the 'Hits' column?

```python
df["Hits"] = df["Hits"].str.replace("\n", "")
df
```
